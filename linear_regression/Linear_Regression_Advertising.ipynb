```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

import seaborn as sns

from sklearn.model_selection import train_test_split

# Supress Warnings
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('advertising.csv')
df.head()
```

```python
%matplotlib inline

sns.pairplot(df, x_vars=['TV', 'Newspaper','Radio'], y_vars='Sales', size=4, kind='scatter')
plt.show()
```

```python
#To plot heatmap to find out correlations

sns.heatmap(df.corr(), cmap='YlGnBu', annot=True)
plt.show()
```

```python
X = df['TV']
y = df['Sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, y, train_size=0.7, test_size=0.3, random_state=100)
```

```python
print (X_train.shape)
print (X_test.shape)
print (y_train.shape)
print (y_test.shape)
```

```python
import statsmodels.api as sm
```

```python
# Add a constant to get an intercept
X_train_sm = sm.add_constant(X_train)

# Fit the resgression line using 'OLS'
lr = sm.OLS(y_train, X_train_sm).fit()
```

```python
# Print the parameters,i.e. intercept and slope of the regression line obtained
lr.params
```

```python
#Performing a summary operation lists out all different parameters of the regression line fitted
print(lr.summary())
```
